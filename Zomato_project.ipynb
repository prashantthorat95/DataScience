## Zomato data analysis project

Step 1 -import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

step 2 - create data  frame

```python
dataframe=pd.read_csv("/content/Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

**convert data type of column = rate**

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

#data cleaning

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('type of restaurant')
```

##conclusion =majority dinning

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurant",c="red",size=15)
plt.ylabel("votes",c="red",size=20)
```

# conclusion =dining has max votes

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

## conclusion= offine gets less rating in compare to online

```python
dataframe.head()
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

## Conclusion: Dining restaurants have the highest count overall, and they are more likely to not offer online ordering compared to other types. Cafes have the second highest count and a significant portion of them offer online ordering. Buffet restaurants have the lowest counts in both categories.
